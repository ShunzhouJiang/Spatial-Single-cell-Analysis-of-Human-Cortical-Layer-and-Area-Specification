```python
import scanpy as sc
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import openpyxl
```

```python
adata = sc.read_h5ad("/Users/cyluo/Library/CloudStorage/Box-Box/MERFISH Collaboration 3/clustering3/h5ad files/gw34_scshc.h5ad")
print(adata)
```

```python
adata_obs = adata.obs
adata_obs
```

```python
merged_encluster_ca_H3 = adata.obs.groupby(['region', 'H1_cluster', 'H2_cluster', 'shc_result']).size().reset_index(name='ca_count')
merged_encluster_ca_H3.to_csv("encluster_ca_H3.csv", index=False)
```

```python
merged_encluster_H3 = adata_obs.groupby(['H1_cluster', 'H2_cluster', 'shc_result']).size().reset_index(name='H3_count')
merged_encluster_H3.to_csv("encluster_H3.csv", index=False)
```

```python
count_ca_H3 = pd.merge(merged_encluster_ca_H3, merged_encluster_H3, on=['H1_cluster', 'H2_cluster', 'shc_result'])
count_ca_H3['percent'] = count_ca_H3['ca_count'] / count_ca_H3['H3_count']
print(count_ca_H3.head)
```

```python
# Check the sum of the 'percent' column for each 'H2_annotation'
percent_sum_check = count_ca_H3.groupby(['H1_cluster', 'H2_cluster', 'shc_result'])['percent'].sum().reset_index(name='percent_sum')

# Display the sums to check if any do not equal 1
print(percent_sum_check)
```
