```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
de_df1 = pd.read_csv('gene_output1.csv')
de_df2 = pd.read_csv('gene_output2.csv')
de_df3 = pd.read_csv('gene_output3.csv')
de_df4 = pd.read_csv('gene_output4.csv')
de_df5 = pd.read_csv('gene_output5.csv')
```

```python
# Sort by log fold change
sorted_de_df1 = de_df1.sort_values(by=['log_fold_change', '-log10_pval'], ascending=[False, False])
sorted_de_df2 = de_df2.sort_values(by=['log_fold_change', '-log10_pval'], ascending=[False, False])
sorted_de_df3 = de_df3.sort_values(by=['log_fold_change', '-log10_pval'], ascending=[False, False])
sorted_de_df4 = de_df4.sort_values(by=['log_fold_change', '-log10_pval'], ascending=[False, False])
sorted_de_df5 = de_df5.sort_values(by=['log_fold_change', '-log10_pval'], ascending=[False, False])
```

```python
# Select top 10/20 and bottom 10/20 genes
N = 20
top_genes1 = sorted_de_df1.head(N)
bottom_genes1 = sorted_de_df1.tail(N)
print("upregulated genes:")
print(top_genes1)
print("downregulated genes:")
print(bottom_genes1)

top_genes2 = sorted_de_df2.head(N)
bottom_genes2 = sorted_de_df2.tail(N)

top_genes3 = sorted_de_df3.head(N)
bottom_genes3 = sorted_de_df3.tail(N)

top_genes4 = sorted_de_df4.head(N)
bottom_genes4 = sorted_de_df4.tail(N)

top_genes5 = sorted_de_df5.head(N)
bottom_genes5 = sorted_de_df5.tail(N)
```

```python
# Combine top and bottom genes
selected_genes1 = pd.concat([top_genes1, bottom_genes1])
selected_genes2 = pd.concat([top_genes2, bottom_genes2])
selected_genes3 = pd.concat([top_genes3, bottom_genes3])
selected_genes4 = pd.concat([top_genes4, bottom_genes4])
selected_genes5 = pd.concat([top_genes5, bottom_genes5])
```

```python
# Assume de_df_example is your dataframe from the differential expression analysis
# Here we replace infinite values with the threshold of 400 for the -log10 p-values
selected_genes1['-log10_pval'] = selected_genes1['-log10_pval'].replace(np.inf, 400)
selected_genes2['-log10_pval'] = selected_genes2['-log10_pval'].replace(np.inf, 400)
selected_genes3['-log10_pval'] = selected_genes3['-log10_pval'].replace(np.inf, 400)
selected_genes4['-log10_pval'] = selected_genes4['-log10_pval'].replace(np.inf, 400)
selected_genes5['-log10_pval'] = selected_genes5['-log10_pval'].replace(np.inf, 400)
```

```python
selected_genes1['color'] = ['red' if x > 0 else 'blue' for x in selected_genes1['log_fold_change']]
plt.figure(figsize=(7, 10))
bar_plot1 = sns.barplot(
    x='log_fold_change',  
    y='gene',  
    hue='color',
    data=selected_genes1,
    dodge=False,
    palette={'red': 'red', 'blue': 'blue'},  
    orient='h'  
)
plt.gca().invert_xaxis()

plt.xlabel('Log Fold Change')
plt.ylabel('Gene')
plt.title('Gene Expression Bar Plot')
bar_plot1.get_legend().remove()

plt.savefig("top20bar_plot1.pdf")

plt.show()
plt.close()

selected_genes2['color'] = ['red' if x > 0 else 'blue' for x in selected_genes2['log_fold_change']]
plt.figure(figsize=(7, 10))
bar_plot2 = sns.barplot(
    x='log_fold_change',  
    y='gene',  
    hue='color',
    data=selected_genes2,
    dodge=False,
    palette={'red': 'red', 'blue': 'blue'},  
    orient='h'  
)
plt.gca().invert_xaxis()

plt.xlabel('Log Fold Change')
plt.ylabel('Gene')
plt.title('Gene Expression Bar Plot')
bar_plot2.get_legend().remove()

plt.savefig("top20bar_plot2.pdf")

plt.show()
plt.close()

selected_genes3['color'] = ['red' if x > 0 else 'blue' for x in selected_genes3['log_fold_change']]
plt.figure(figsize=(7, 10))
bar_plot3 = sns.barplot(
    x='log_fold_change',  
    y='gene',  
    hue='color',
    data=selected_genes3,
    dodge=False,
    palette={'red': 'red', 'blue': 'blue'},  
    orient='h'  
)
plt.gca().invert_xaxis()

plt.xlabel('Log Fold Change')
plt.ylabel('Gene')
plt.title('Gene Expression Bar Plot')
bar_plot3.get_legend().remove()

plt.savefig("top20bar_plot3.pdf")

plt.show()
plt.close()

selected_genes4['color'] = ['red' if x > 0 else 'blue' for x in selected_genes4['log_fold_change']]
plt.figure(figsize=(7, 10))
bar_plot4 = sns.barplot(
    x='log_fold_change',  
    y='gene',  
    hue='color',
    data=selected_genes4,
    dodge=False,
    palette={'red': 'red', 'blue': 'blue'},  
    orient='h'  
)
plt.gca().invert_xaxis()

plt.xlabel('Log Fold Change')
plt.ylabel('Gene')
plt.title('Gene Expression Bar Plot')
bar_plot4.get_legend().remove()

plt.savefig("top20bar_plot4.pdf")

plt.show()
plt.close()

selected_genes5['color'] = ['red' if x > 0 else 'blue' for x in selected_genes5['log_fold_change']]
plt.figure(figsize=(7, 10))
bar_plot5 = sns.barplot(
    x='log_fold_change',  
    y='gene',  
    hue='color',
    data=selected_genes5,
    dodge=False,
    palette={'red': 'red', 'blue': 'blue'},  
    orient='h'  
)
plt.gca().invert_xaxis()

plt.xlabel('Log Fold Change')
plt.ylabel('Gene')
plt.title('Gene Expression Bar Plot')
bar_plot1.get_legend().remove()

plt.savefig("top20bar_plot5.pdf")

plt.show()
plt.close()
```
