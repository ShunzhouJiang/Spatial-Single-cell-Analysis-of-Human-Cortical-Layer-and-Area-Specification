```python
import scanpy as sc
import pandas as pd
import matplotlib.pyplot as plt
```

```python
adata = sc.read_h5ad("/Users/cyluo/Library/CloudStorage/Box-Box/MERFISH Collaboration 3/clustering3/h5ad files/zenodo/merscope_integrated_855.h5ad")
adata.obs
```

```python
filtered_adata = adata[adata.obs['H1_annotation'].isin(["EN-IT", "EN-ET", "EN-Mig"])]
```

```python
cortical_area = pd.read_excel("cortical_area.xlsx")
print(cortical_area.head())
```

```python
encluster_df = filtered_adata.obs
encluster_age = encluster_df.groupby('gw', observed=False).size().reset_index(name='age_count')
encluster_age.to_csv("encluster_age.csv", index=False)
```

```python
encluster_age['gw'] = pd.Categorical(encluster_age['gw'], categories=["15", "20", "22", "34"], ordered=True)
```

```python
labels = encluster_age['gw']
sizes = encluster_age['age_count']
colors = ["#B0F26D", "#5ebe5a", "#F066EA", "blueviolet"]
```

```python
plt.figure(figsize=(8, 8)) 
plt.pie(
    sizes,
    labels=labels,
    colors=colors,
    autopct='%1.1f%%', 
    startangle=0,
    wedgeprops={"edgecolor": "white"}, 
)
```
