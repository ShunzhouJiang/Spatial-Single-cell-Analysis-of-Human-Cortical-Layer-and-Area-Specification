```python
import scanpy as sc
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# mers_imputed_1000.h5ad is generated by EDFig18/EDFig18A_ENVI_imputation.py
adata = sc.read_h5ad("/Users/cyluo/Library/CloudStorage/Box-Box/MERFISH Collaboration 3/clustering3/h5ad files/zenodo/mers_imputed_1000.h5ad")
print(adata)
```

```python
adata_gw20 = sc.read_h5ad("/Users/cyluo/Library/CloudStorage/Box-Box/MERFISH Collaboration 3/clustering3/h5ad files/zenodo/gw20.h5ad")
print(adata_gw20)
```

```python
print(adata_gw20.obs.head())
```

```python
adata_impute = adata.obsm['imputation']
adata_impute = adata_impute[list(set(adata_impute.columns) - set(adata.var.index))]
adata_impute = sc.AnnData(adata_impute)
adata_tot = sc.concat([adata, adata_impute], axis=1)
adata_tot.obs = adata.obs
```

```python
print(adata_tot.obs.head())
```

```python
adata_subset = adata_tot[adata_tot.obs['gw'] == 20].copy()
print(adata_subset.obs.head())
```

```python
adata_gw20.obs['cell_id'] = adata_gw20.obs.index.str.split('-').str[0].astype(str)
```

```python
adata_subset.obs['cell_id'] = adata_subset.obs.index.astype(str)
```

```python
adata_subset.obs = pd.merge(
    adata_subset.obs, 
    adata_gw20.obs[['cell_id', 'H3_annotation']],  
    on='cell_id', 
    how='left'
)
print(adata_subset.obs)
```

```python
adata_subset.obs['H3_annotation'].unique()
```

```python
clusters_of_interest1 = ["EN-IT-L3-c0", "EN-IT-L3-c3"]
clusters_of_interest2 = ["EN-IT-L3/4-c3", "EN-IT-L3/4-c0"]
clusters_of_interest3 = ["EN-ET-L5/6-c4", "EN-ET-L5/6-c3"]
clusters_of_interest4 = ['EN-ET-SP-2-c1', 'EN-ET-SP-2-c3']
adata_subset1 = adata_subset[adata_subset.obs['H3_annotation'].isin(clusters_of_interest1)]
adata_subset2 = adata_subset[adata_subset.obs['H3_annotation'].isin(clusters_of_interest2)]
adata_subset3 = adata_subset[adata_subset.obs['H3_annotation'].isin(clusters_of_interest3)]
adata_subset4 = adata_subset[adata_subset.obs['H3_annotation'].isin(clusters_of_interest4)]
print(adata_subset1.obs)
```

```python
# Perform DE analysis
sc.tl.rank_genes_groups(adata_subset1, groupby='H3_annotation') 
sc.tl.rank_genes_groups(adata_subset2, groupby='H3_annotation') 
sc.tl.rank_genes_groups(adata_subset3, groupby='H3_annotation') 
sc.tl.rank_genes_groups(adata_subset4, groupby='H3_annotation') 
```

```python
result1 = adata_subset1.uns['rank_genes_groups']
result2 = adata_subset2.uns['rank_genes_groups']
result3 = adata_subset3.uns['rank_genes_groups']
result4 = adata_subset4.uns['rank_genes_groups']

de_df1 = pd.DataFrame({
    'gene': result1['names']['EN-IT-L3-c0'], 
    'log_fold_change': result1['logfoldchanges']['EN-IT-L3-c0'],
    'pvals': result1['pvals']['EN-IT-L3-c0'],
    'pvals_adj': result1['pvals_adj']['EN-IT-L3-c0']
})
de_df1['-log10_pval'] = -np.log10(de_df1['pvals_adj'])

de_df2 = pd.DataFrame({
    'gene': result2['names']['EN-IT-L3/4-c3'], 
    'log_fold_change': result2['logfoldchanges']['EN-IT-L3/4-c3'],
    'pvals': result2['pvals']['EN-IT-L3/4-c3'],
    'pvals_adj': result2['pvals_adj']['EN-IT-L3/4-c3']
})
de_df2['-log10_pval'] = -np.log10(de_df2['pvals_adj'])

de_df3 = pd.DataFrame({
    'gene': result3['names']['EN-ET-L5/6-c4'], 
    'log_fold_change': result3['logfoldchanges']['EN-ET-L5/6-c4'],
    'pvals': result3['pvals']['EN-ET-L5/6-c4'],
    'pvals_adj': result3['pvals_adj']['EN-ET-L5/6-c4']
})
de_df3['-log10_pval'] = -np.log10(de_df1['pvals_adj'])

de_df4 = pd.DataFrame({
    'gene': result4['names']['EN-ET-SP-2-c1'], 
    'log_fold_change': result4['logfoldchanges']['EN-ET-SP-2-c1'],
    'pvals': result4['pvals']['EN-ET-SP-2-c1'],
    'pvals_adj': result4['pvals_adj']['EN-ET-SP-2-c1']
})
de_df4['-log10_pval'] = -np.log10(de_df4['pvals_adj'])
```

```python
de_df1.to_csv('gene1000_output1.csv', index = False)
de_df2.to_csv('gene1000_output2.csv', index = False)
de_df3.to_csv('gene1000_output3.csv', index = False)
de_df4.to_csv('gene1000_output4.csv', index = False)
```
