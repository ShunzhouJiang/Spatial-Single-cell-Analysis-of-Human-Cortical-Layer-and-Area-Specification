```python
import os
import matplotlib.pyplot as plt
import scanpy as sc
import numpy as np
import pandas as pd
# from multiprocessing import Pool
from itertools import repeat
import warnings
import sys
from anndata import read_h5ad, concat
warnings.filterwarnings('ignore')
from matplotlib.colors import TwoSlopeNorm

from Fig6DEF_DEG_time_detect import *
from Fig6DEF_DEG_time_plot import *
```

```python
gw_num_lst = ["15",  "20",  "34", "Adult"]
v1v2_type_lst = ["v1", "v2"]
itet_lst = ["IT", "ET"]
deg_save_path = "DEG_time_result"

for v1v2_type in v1v2_type_lst:
    for itet in itet_lst:
        for i in range(len(gw_num_lst)):
            de_df = find_deg(gw_num_lst[i], v1v2_type, itet = itet, path = deg_save_path)
            print(de_df[(de_df['log_fold_change'] > 0.5) & (de_df['pvals_adj'] <= 0.05)].shape[0])
```

```python
deg_plot_path = "DEG_time_plot"
v1v2_type_lst = ["v1", "v2"]
itet_lst = ["IT", "ET"]
top_num_lst = [False]


for v1v2_type in v1v2_type_lst:
    for itet in itet_lst:
        for top_num in top_num_lst:
            if top_num:
                plot_inv_gene = False
            else:
                plot_inv_gene = True
            lfc_tot = plot_tot(itet = f"EN-{itet}", v1v2_type = v1v2_type.upper(), top_num = top_num, plot_inv_gene = plot_inv_gene, 
                            deg_path = deg_save_path, deg_plot_path = deg_plot_path)

            lfc_tot = np.array(lfc_tot)
            gw_lst = ["15", "20", "34", "Adult"]

            if v1v2_type.upper() == "V1":
                cmap = plt.cm.get_cmap("PiYG_r")
            else:
                cmap = plt.cm.get_cmap("PiYG")
            # cmap = plt.cm.get_cmap("PiYG")
            norm = TwoSlopeNorm(vmin=-0.4, vcenter=0, vmax=0.8) 

            os.makedirs(deg_plot_path, exist_ok = True)
            fig = plt.figure(figsize=(6,5))
            sns.heatmap(
                lfc_tot,
                cmap=cmap,
                norm=norm,
                cbar_kws={"label": "Log fold change"}, 
                vmin=-0.4, 
                vmax=0.8,
                annot=False 
            )

            # Add x and y labels
            plt.xticks(ticks=np.arange(lfc_tot.shape[1]) + 0.5, labels=gw_lst, rotation=45)
            plt.yticks(ticks=np.arange(lfc_tot.shape[0]) + 0.5, labels=gw_lst)

            if top_num:
                fig.savefig(f"{deg_plot_path}/{itet}_{v1v2_type.upper()}_top{top_num}_heatmap.pdf", format="pdf", dpi = 300)
            else:
                fig.savefig(f"{deg_plot_path}/{itet}_{v1v2_type.upper()}_all_heatmap.pdf", format="pdf", dpi = 300)
            # plt.show()
            plt.close()
            
            lfc_tot = pd.DataFrame(lfc_tot)
            lfc_tot.index = gw_lst
            lfc_tot.columns = gw_lst
            if top_num:
                lfc_tot.to_csv(f"{deg_plot_path}/{itet}_{v1v2_type.upper()}_top{top_num}_lfc.csv")
            else:
                lfc_tot.to_csv(f"{deg_plot_path}/{itet}_{v1v2_type.upper()}_all_lfc.csv")
```
